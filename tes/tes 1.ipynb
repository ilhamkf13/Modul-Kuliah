```python
# Fungsi untuk menghitung total biaya
def hitung_biaya(harga_kamar, lama_menginap, member):
    if member == 'Y':
        diskon = 0.2  # Diskon 20% untuk member
    else:
        diskon = 0
    
    total_biaya = harga_kamar * lama_menginap * (1 - diskon)
    return total_biaya

# Input data dari tamu
nama_tamu = input("Masukkan nama tamu: ")
alamat_tamu = input("Masukkan alamat tamu: ")

print("Pilihan keanggotaan member hotel (Y/N): ")
keanggotaan_member = input()
while keanggotaan_member not in ['Y', 'N']:
    print("Pilihan tidak valid. Harap masukkan 'Y' atau 'N'.")
    keanggotaan_member = input()

print("Pilihan jenis kamar:")
print("1. Suite (Harga kamar Rp 1.000.000)")
print("2. Superior (Harga kamar Rp 600.000)")
print("3. Deluxe (Harga kamar Rp 400.000)")
jenis_kamar = int(input("Masukkan pilihan jenis kamar (1/2/3): "))
while jenis_kamar not in [1, 2, 3]:
    print("Pilihan tidak valid. Harap masukkan 1, 2, atau 3.")
    jenis_kamar = int(input())

# Tentukan harga kamar berdasarkan pilihan
if jenis_kamar == 1:
    harga_kamar = 1000000
elif jenis_kamar == 2:
    harga_kamar = 600000
else:
    harga_kamar = 400000

lama_menginap = int(input("Masukkan lama menginap (dalam hari): "))

# Hitung total biaya
total_biaya = hitung_biaya(harga_kamar, lama_menginap, keanggotaan_member)

# Tampilkan data dan total biaya
print("\nData Tamu:")
print("Nama Tamu:", nama_tamu)
print("Alamat Tamu:", alamat_tamu)
print("Keanggotaan Member Hotel:", keanggotaan_member)
print("Jenis Kamar:", jenis_kamar)
print("Harga Kamar per Hari:", harga_kamar)
print("Lama Menginap:", lama_menginap, "hari")
print("Total Biaya yang Harus Dibayarkan:", "Rp", total_biaya)
```
